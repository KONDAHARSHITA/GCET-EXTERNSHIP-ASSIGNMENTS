```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset["Genre"]=le.fit_transform(dataset["Genre"])
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,0:4].values
y=dataset.iloc[:,4].values
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler 
sc =StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x.shape
```

```python
dataset.head()
```

```python
from tensorflow.keras.models import Sequential #-used to initialize the model.
from tensorflow.keras.layers import Dense #- used to add layers of ANN
```

```python
con=Sequential()
```

```python
con.add(Dense(4,kernel_initializer="random_uniform",activation='relu'))
```

```python
con.add(Dense(8,kernel_initializer="random_uniform",activation='relu'))
```

```python
con.add(Dense(1,kernel_initializer="random_uniform",activation='relu'))
```

```python
con.compile("adam","mean_squared_error", metrics=['mse'])
```

```python
con.fit(x_train,y_train,batch_size=32,epochs=10)
```

```python
ypred=con.predict(x_test)
```
